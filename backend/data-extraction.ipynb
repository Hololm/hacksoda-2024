```python
def get_sellers_data(sellers: dict[str]):
    data = []
    
    with ThreadPoolExecutor() as executor:
        futures = [executor.submit(get_seller, seller) for seller in sellers]
        
        for future in as_completed(futures):
            try:
                data.append(future.result())
            except Exception as e:
                print(e)
    
    return data
```
